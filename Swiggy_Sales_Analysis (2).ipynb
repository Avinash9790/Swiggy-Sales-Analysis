**Charts**

```python
# Monthly Trend
Swiggy['Order Date']=pd.to_datetime(Swiggy['Order Date'])
Swiggy['YearMonth'] = Swiggy['Order Date'].dt.to_period('M').astype(str)
Monthly_Revenue = Swiggy.groupby('YearMonth')['Price (INR)'].sum()

plt.figure(figsize=(12, 6))
plt.plot(Monthly_Revenue.index, Monthly_Revenue.values)
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.title('Monthly Trend')
plt.xticks(rotation=45)
```

```python
# Daily Trend
Swiggy['Date'] = Swiggy['Order Date'].dt.date
Swiggy['Day Name'] = Swiggy['Order Date'].dt.day_name()

day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

daily_revenue = Swiggy.groupby('Day Name')['Price (INR)'].sum().reindex(day_order)

plt.figure(figsize=(12, 6))
bars = plt.bar(daily_revenue.index, daily_revenue.values)
plt.xlabel('Day of the Week')
plt.ylabel('Revenue')
plt.title('Daily Trend')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')
```

```python
# Revenue by State
state_revenue = Swiggy.groupby('State', as_index= False)['Price (INR)'].sum().sort_values(by='Price (INR)', ascending= True)

plt.figure(figsize=(25, 12))
bar2 = plt.barh(state_revenue['State'], state_revenue['Price (INR)'])
plt.xlabel('Revenue')
plt.ylabel('State')
plt.title('Revenue by State')

for bar in bar2:
    width = bar.get_width()  # This gets the Revenue value
    label_y = bar.get_y() + bar.get_height() / 2  # This finds the center of the bar

    plt.text(width, label_y, f' {width:,.0f}', va='center')
```

```python
# Revnue by Food Category

non_veg_keywords = [
    'egg', 'chicken', 'fish', 'mutton', 'prawn',
    'kebab', 'kabab','biryani', 'non-veg', 'non veg'
]

Swiggy['Food Category'] = np.where(
    Swiggy['Dish Name'].str.lower().str.contains('|'.join(non_veg_keywords)),
    'Non-Veg',
    'Veg'
)

food_category_revenue = Swiggy.groupby('Food Category')['Price (INR)'].sum().reset_index()

ax = px.pie(food_category_revenue, names='Food Category', values='Price (INR)', hole= 0.5, title='Revenue by Food Category')
ax.update_traces(textposition='inside', textinfo='percent+label')
ax.show()
```

```python
# Quarterly Trend

quarterly_summary = (Swiggy.assign(
    Order_Date=pd.to_datetime(Swiggy["Order Date"]),
    Quarter=lambda x: x["Order_Date"].dt.to_period("Q").astype(str)
)
.groupby("Quarter", as_index=False)
.agg(
    Total_Sales=("Price (INR)", "sum"),
    Avg_Rating=("Rating", "mean"),
    Total_Orders=("Order_Date", "count")
)
.sort_values("Quarter")
)

quarterly_summary["Total_Sales"] = quarterly_summary["Total_Sales"].round(0)
quarterly_summary["Avg_Rating"] = quarterly_summary["Avg_Rating"].round(2)

quarterly_summary
```

```python
# Top 5 cities

top_5_cities = Swiggy.groupby('City')['Price (INR)'].sum().nlargest(5).sort_values().reset_index()

ax2 = px.bar(top_5_cities, x='Price (INR)', y='City', title='Top 5 Cities by Revenue')
ax2.update_layout(xaxis_title='Revenue', yaxis_title='City')
ax2.show()
```

```python
# Weekly Trend

Swiggy['Week Number'] = Swiggy['Order Date'].dt.isocalendar().week

weekly_revenue = Swiggy.groupby('Week Number')['Price (INR)'].sum().reset_index()

ax3 = px.bar(weekly_revenue, x='Week Number', y='Price (INR)', title='Weekly Revenue Trend')
ax3.update_layout(xaxis_title='Week Number', yaxis_title='Revenue')
ax3.show()
```

```python
# Top Cities by Orders

plt.figure(figsize=(12,6))
top_cities = Swiggy['City'].value_counts().head(10)
sns.barplot(x=top_cities.values, y=top_cities.index, palette='Reds_r')
plt.title('Top 10 Cities by Number of Orders')
plt.xlabel('Number of Orders')
plt.ylabel('City')
plt.tight_layout()
plt.show()
```

```python
# Top 10 most ordered dishes

plt.figure(figsize=(12,6))
top_dishes = Swiggy['Dish Name'].value_counts().head(10)
sns.barplot(x=top_dishes.values, y=top_dishes.index, palette='Reds_r')
plt.title('Top 10 Most Ordered Dishes')
plt.xlabel('Number of Orders')
plt.ylabel('Dish Name')
plt.tight_layout()
plt.show()
```

```python
# Top Food Categories:

plt.figure(figsize=(12,6))
top_categories = Swiggy['Category'].value_counts().head(10)
sns.barplot(x=top_categories.values, y=top_categories.index, palette='Reds_r')
plt.title('Top 10 Food Categories by Orders')
plt.xlabel('Number of Orders')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

```python
# Monthly Orders Trend

Swiggy['Order Date'] = pd.to_datetime(Swiggy['Order Date'])
Swiggy['Month'] = Swiggy['Order Date'].dt.to_period('M')
monthly_orders = Swiggy.groupby('Month').size()
plt.figure(figsize=(12,6))
monthly_orders.plot(kind='line', marker='o', color='red')
plt.title('Monthly Orders Trend')
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.tight_layout()
plt.show()
```

```python
# Price Distribution

plt.figure(figsize=(12,6))
sns.histplot(Swiggy['Price (INR)'], bins=50, color='red', kde=True)
plt.title('Price Distribution')
plt.xlabel('Price (INR)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Rating Distribution

plt.figure(figsize=(12,6))
sns.histplot(Swiggy['Rating'], bins=30, color='darkred', kde=True)
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
